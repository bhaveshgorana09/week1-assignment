###  The Geometry of Matrices: Understanding Eigenvalues & Eigenvectors

Think of a **Matrix** as a linear transformation machine. When you apply it to a vector space, its generic action is to warp the environment by **rotating** and **scaling** almost every vector within it.

However, during this transformation, **Eigenvectors** serve as the anchor lines of the space:

>  **Eigenvectors** are special, invariant directions that completely resist rotation. When multiplied by the matrix, they never tilt, turn, or leave their path; they stay perfectly locked onto their original linear track (span).


### The Scaling Factor: Eigenvalues ($\lambda$)

An **Eigenvalue ($\lambda$)** is the precise scalar multiplier that dictates how the corresponding eigenvector behaves along its invariant track:

*  **Stretching ($|\lambda| > 1$):** The matrix drives the eigenvector outward away from the origin, increasing its **length**.
  * *In your code:* The first eigenvector is stretched to exactly **5 times** its original length because $\lambda = 5$.
*  **Squishing ($0 < |\lambda| < 1$):** The matrix compresses the eigenvector inward toward the origin, making its length **shorter**.
*  **Flipping ($\lambda < 0$):** The eigenvector stays on the exact same linear path, but it gets pulled backward through the origin to point in the **opposite direction**.

### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).
2. Compute its **SVD**: `U, S, Vt = np.linalg.svd(X, full_matrices=False)`.
3. **Reconstruct** `X` from the SVD and verify it matches the original.
4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````python
np.random.seed(42)
X = np.random.randn(4, 3)

U, S, Vt = np.linalg.svd(X, full_matrices=False)

X_reconstructed = U @ np.diag(S) @ Vt
X_approx        = U[:, :1] @ np.diag(S[:1]) @ Vt[:1, :]

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"
print("4.4 passed ✓")
````

**SVD → PCA Connection (4.4):**

> SVD breaks down a data matrix into three distinct parts: $U$, $S$, and $Vt$. When your data is mean-centered, these components map directly to PCA:

* $U$ (Left Singular Vectors): These represent the normalized projections of your original data points onto the principal components. They tell you the position or coordinates of your data in the new feature space.

* $S$ (Singular Values): This is a diagonal matrix where the values represent the strength or significance of each axis. The square of these singular values is directly proportional to the variance (eigenvalues) captured by each principal component.

* $Vt$ (Right Singular Vectors): This matrix contains the directions of maximum variance. The rows of $Vt$ are the actual Principal Components (or loadings).


**Matrix containing the directions of maximum variance:**
The matrix **$V^T$** (specifically, its first row) contains the principal axes or direction of maximum variance. When you project your centered data matrix $X$ onto these vectors, the first column of $U \cdot S$ reveals the coordinates along the path where your dataset is most widely distributed.

### 5.4 Distribution Testing & Stationarity

1. Generate two samples: `s1` ~ Normal(0, 1), `s2` ~ Exponential(1), both n=200 (seed=0).
2. Run the **Kolmogorov-Smirnov test** (`stats.kstest`) comparing each against a standard normal CDF.
3. Generate a non-stationary time series (linear trend + noise). Run the **Augmented Dickey-Fuller test** (`adfuller` from `statsmodels`) and interpret the result.
4. Difference the series once and re-run ADF to confirm stationarity.

````python
from statsmodels.tsa.stattools import adfuller

np.random.seed(0)
s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# KS test — compare each against N(0,1)
ks_s1 = stats.kstest(s1, 'norm')
ks_s2 = stats.kstest(s2, 'norm')
print(f"KS test s1 (normal):      stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

# ADF test on non-stationary series
t = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)

adf_result = adfuller(ts)
print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value:   {adf_result[1]:.4f}")
print(f"Stationary?    {adf_result[1] < 0.05}")

# Difference once and re-run
ts_diff = np.diff(ts)
adf_diff = adfuller(ts_diff)
print(f"\nAfter differencing — p-value: {adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}")
````

### 5.5 Model Monitoring Concepts

1. Implement `compute_psi(expected, actual, bins=10)` that computes the **Population Stability Index** between two distributions.
2. Simulate a scenario where a model's input distribution shifts and plot both distributions.
3. In the markdown cell, define: *concept drift*, *covariate drift*, PSI thresholds (what values indicate no/minor/major shift), and one trigger for retraining.

````python
def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    bin_edges = np.percentile(expected, np.linspace(0, 100, bins + 1))
    bin_edges[0] -= 1e-5
    bin_edges[-1] += 1e-5

    expected_counts, _ = np.histogram(expected, bins=bin_edges)
    actual_counts, _ = np.histogram(actual, bins=bin_edges)

    expected_pct = expected_counts / len(expected)
    actual_pct = actual_counts / len(actual)

    eps = 1e-10
    expected_pct = np.where(expected_pct == 0, eps, expected_pct)
    actual_pct = np.where(actual_pct == 0, eps, actual_pct)

    psi_value = np.sum((actual_pct - expected_pct) * np.log(actual_pct / expected_pct))
    return psi_value
    pass

np.random.seed(1)
train_dist = np.random.normal(50, 10, 1000)    # training distribution
drift_dist = np.random.normal(65, 12, 1000)    # shifted production distribution

psi_value = compute_psi(train_dist, drift_dist)
print(f"PSI: {psi_value:.4f}")
print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")

# Plot both distributions
fig, ax = plt.subplots()
sns.histplot(train_dist, kde=True, label='Training Distribution', color='blue', alpha=0.5, ax=ax)
sns.histplot(drift_dist, kde=True, label='Shifted Production', color='red', alpha=0.5, ax=ax)
ax.legend(); ax.set_title('Distribution Shift Visualization')
plt.tight_layout(); plt.show()
````

**Concept Drift vs Covariate Drift & PSI Thresholds (5.5):**

- **Concept drift:** A change in the statistical properties of the target variable over time, meaning the relationship between the input features and the actual labels has changed ($P(Y|X)$ changes).
- **Covariate drift:** A shift in the distribution of the input data over time while the underlying relationship between features and labels remains the same ($P(X)$ changes, as seen in your visualization).
- **PSI < 0.1:** **Stable / No Shift;** The production data distribution is significantly identical to the baseline/training distribution, requiring no action.
- **PSI 0.1–0.2:** **Moderate Shift;** The data distribution has drifted slightly; it is a warning threshold where you should monitor performance closely or plan a future update.
- **PSI > 0.2:** **Major Shift;** The distribution has changed significantly (matching your plot's "Major" severity), indicating the model's environment has shifted drastically.

### 6.1 Core Concepts

A bag contains: 4 red, 3 blue, 3 green marbles.

1. Define the sample space and compute P(red), P(blue), P(green).
2. Two marbles drawn **without replacement**. Compute:
   - P(first=red, second=blue) — **joint probability**
   - P(second=blue | first=red) — **conditional probability**
3. Are the two draws **independent**? Show mathematically.

````python
total = 10
red, blue, green = 4, 3, 3

p_red = red / total
p_blue = blue / total
p_green = green / total

# Joint: P(red first AND blue second)
p_red_then_blue = (red / total) * (blue / (total - 1))

# Conditional Probability: P(second=blue | first=red)
p_blue_given_red = blue / (total - 1)

# Independence check: are P(B|A) == P(B) ?
independent = p_blue_given_red == p_blue

print(f"P(red)={p_red:.3f}, P(blue)={p_blue:.3f}, P(green)={p_green:.3f}")
print(f"P(red,blue): {p_red_then_blue:.4f}")
print(f"P(blue|red): {p_blue_given_red:.4f}")
print(f"P(blue):     {p_blue:.4f}")
print(f"Independent? {independent}")
````

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

````python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# P(free) — law of total probability
p_free = (p_free_given_spam * p_spam) + (p_free_given_ham * p_ham)

# Bayes
p_spam_given_free = (p_free_given_spam * p_spam) / p_free


print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    prior_ham = 1 - prior_spam
    p_word = (p_word_given_spam * prior_spam) + (p_word_given_ham * prior_ham)
    posterior_spam = (p_word_given_spam * prior_spam) / p_word
    return posterior_spam
    pass


pred = naive_bayes_predict(0.30, 0.80, 0.05)
assert abs(pred - p_spam_given_free) < 1e-9
print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
````

**Bayes Term Mapping (6.3):**

| Term | Formula | In this example |
|------|---------|-----------------|
| Prior | P(Spam) | The initial probability that any incoming email is spam before looking at its text (**0.3000** or 30%). |
| Likelihood | P(word \| Spam) | The probability that the word "free" appears, given that the email is already known to be spam (**0.8000** or 80%). |
| Evidence | P(word) | The total probability of encountering the word "free" across all incoming emails, whether ham or spam (**0.2750** or 27.5%). |

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
np.random.seed(7)
population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30
sample_means = np.array([np.random.choice(population, size=sample_size).mean()
for _ in range(n_samples)])


pop_mean = population.mean()
pop_std  = population.std()
clt_std  = pop_std / np.sqrt(sample_size)

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Plot
fig, ax = plt.subplots()
ax.hist(sample_means, bins=30, density=True, alpha=0.6)
x = np.linspace(sample_means.min(), sample_means.max(), 300)

ax.plot(x, stats.norm.pdf(x, pop_mean, clt_std))
ax.set_title('CLT: Distribution of Sample Means')
plt.tight_layout(); plt.show()

# KS test
ks_result = stats.kstest(sample_means, 'norm', args=(pop_mean, clt_std))
print(f"KS test p-value: {ks_result.pvalue:.4f} → Approximately normal? {ks_result.pvalue > 0.05}")
````

**CLT Reflection (6.4):**

> The Central Limit Theorem (CLT) is vital because it guarantees that as long as our sample size is sufficiently large, the sample means will form a normal distribution regardless of how skewed or non-normal the underlying population data is. This behavior justifies the fundamental **assumption of normality** required by many classical statistical tests, such as t-tests and ANOVA. In machine learning, this allows us to safely calculate reliable confidence intervals, perform parametric hypothesis tests, and assume normally distributed residual errors when evaluating model parameters.
